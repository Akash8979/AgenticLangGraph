## Schema for Structured Output

```python
from langchain_groq import ChatGroq
from pydantic import BaseModel,Field
from dotenv import load_dotenv
load_dotenv()

llm = ChatGroq(model='qwen/qwen3-32b')

class SearchQeary(BaseModel):
    search_query: str=Field(None,description='Qeary that is optimized search.')
    justification:str=Field(None,description="Why this query is relevent to the user's request.")

# Augment the llm with the schema of structure output
structured_llm = llm.with_structured_output(SearchQeary)

# output = structured_llm.invoke("Why is sleep import ?")
# print(output)
```

## Tool binding

```python
llm = ChatGroq(model='qwen/qwen3-32b')

def multiply(a:int,b:int)->int:
    return a*b

llm_with_tools = llm.bind_tools([multiply])

llm_output = llm_with_tools.invoke("what is multiply 3 and 7 ?")
print(llm_output.tool_calls)
```
