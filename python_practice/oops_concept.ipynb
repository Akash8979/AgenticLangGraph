```python
# create class

class Employee:
    def __init__(self,employee_name):
        self.name = employee_name

    def introduce(self):
        print("Hello I am " + self.name)

employee_obj = Employee("Akash")
print(employee_obj.name)
employee_obj.introduce()
```

```python
# inheritance in python oops
class ParentClass:
    def par_func(self):
        print("I am parent class function")

class ChildClass(ParentClass):
    def child_func(self):
        print("I am child class")

obj1 = ChildClass()
obj1.par_func()
obj1.child_func()
```

```python
# Multi level inheritance

class A:
    def __init__(self,a_name):
        self.a_name = a_name

class B(A):
    def __init__(self,b_name,a_name):
        self.b_name=b_name
        A.__init__(self,a_name)

class C(B):
    def __init__(self,c_name,b_name,a_name):
        self.c_name =c_name
        B.__init__(self,b_name,a_name)  

    def display_name(self):
        print("A Name",self.a_name)     
        print("B Name",self.b_name)     
        print("C Name",self.c_name)     

obj_2 = C("c","b","a")  
print(obj_2.a_name)
obj_2.display_name()
```

```python
# multiple Inheritance

class Parent1:
    def parent1_func(self):
        print("Hi i am fist parent function")

class Parent2:
    def parent2_func(self):
        print("Hi i am second parent function")

class Child(Parent1,Parent2):
    def child_func(self):
        self.parent1_func()
        self.parent2_func()

obj1 = Child()
obj1.child_func()

```

```python
# Hierarchical inheritance: when a parent class is drived by more than one child class. it is called hierachical inheritance.
class A:
    def a_func(self):
        print("Hi i am parent class")

class B(A):
    def b_func(self):
        print("I am from the first child")

class C(A):
    def c_func(self):
        print("I am from the second child")

obj1 = B()
obj2 = C()
obj1.a_func()
obj1.b_func()
obj2.c_func()
obj2.a_func()

```

```python
# access parent members using Parent class name:

class Parent:
    def __init__(self,name):
        self.name = name

class Child(Parent):
    def __init__(self,name,age):
        Parent.name = name
        self.age = age

    def display(self):
        print(Parent.name,self.age)


obj = Child("Interview bit",6)         
obj.display()   
```

```python
# access parent members using super():
class Parent:
    def __init__(self,name):
        self.name = name

class Child(Parent):
    def __init__(self,name,age):
        super(Child,self).__init__(name)
        self.age = age
    def display(self):    
        print(self.name,self.age)

obj = Child("InterviewBit",6)
obj.display()
```
