```python
# brute force approach to find pair sum
A = [-1, 5, 7, 6, 5]
K = 6

s = len(A)

for i in range(0,s):
    for j in range(i+1,s):
        if A[i] + A[j] == K:
            print((i,j))
```

```python
# Two point approach to find pair sum
A = [-1, 5, 7, 6, 5]
K = 6

l = 0
r = len(A) - 1

while l < r:
    s = A[r] - A[l]
    if A[l] + A[r] == K:
        print((A[l],A[r]))
        l +=1
        r -=1
    elif s < K:
        l +=1
    else:
        r -=1            
```

```python
#  hesh set appraoch to find pair sum
A = [1, 5, 7, -1, 5]
K = 6

seen = set()
output = set()
for item in A:
    p = (K - item)
    if p in seen:
        output.add((p,item))
    seen.add(item)    

print(output)
```

```python
#  hesh set appraoch to find pair sum 
A = [1, 5, 7, -1, 5]
K = 6

seen = {}
for i, item in enumerate(A):
    p = (K - item)
    if p in seen:
        print(seen[p],i)
    seen[item] = i   

```

```python
#A Program to add two integers >0 without using the plus operator.

def add_num(num1,num2):
    while num2 != 0:
        num1, num2 = num1 ^ num2, (num1 & num2) << 1
    print(num1)    
add_num(8,4)    
```

```python
#A Program to match a string that has the letter ‘a’ followed by 4 to 8 'b’s.
import re

def match_text(s):
    pattern = "ab{4,8}"
    if re.search(pattern,s):
        print("Match Found")
    else:
        print("Match not found")   

match_text("abc")
match_text("aabbbbbbbb")
```

```python
#  Program to combine two different dictionaries. While combining, 
# if you find the same keys, you can add the values of these same keys. Output the new dictionary

from collections import Counter

d1 = {'key1': 50, 'key2': 100, 'key3':200}
d2 = {'key1': 200, 'key2': 100, 'key4':300}
new_dict = Counter(d1) + Counter(d2)
print(new_dict)
```

```python
import threading
import time

def count(n):
    while n > 0:
        n -= 1

start = time.time()
t1 = threading.Thread(target=count, args=(50_000_000,))
t2 = threading.Thread(target=count, args=(50_000_000,))
t1.start(); t2.start()
t1.join(); t2.join()
print(f"Threaded: {time.time() - start:.2f}s")

start = time.time()
count(50_000_000)
count(50_000_000)
print(f"Sequential: {time.time() - start:.2f}s")
```

```python
def is_valid(s):
    stack = []
    pairs = {
        ")":"(",
        "}":"{",
        "]":"["
    }
    for ch in s:
        if ch in pairs.values():
            stack.append(ch)
        elif ch in pairs and (not stack or stack.pop() != pairs[ch]):
            return False
    return  not stack

print(is_valid("()[]{}"))   # True
print(is_valid("(]"))       # False
```
