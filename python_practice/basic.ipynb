```python
### Basic Dacorator and basic calling
def decorator_function(original_function):
    def wrapper():
        print("before")
        original_function()
        print("after")
    return wrapper

def display():
    print("hello world")

# here it is decorated
decorated = decorator_function(display)    
decorated()

```

```python
# Basic Decorator with @ sytext call

def decorator_function(original_function):
    def wrapper():
        print("Before original fucntion call")
        original_function()
        print("After origin function call")
    return wrapper

@decorator_function
def display():
    print("hello world")

display()

```

```python
# Decorator with arguments

def decorator_function(original_function):
    def wrapper(*args,**kwargs):
        result  = original_function(*args,**kwargs)
        return result
    return wrapper

@decorator_function
def add(a,b):
    return a + b

print(add(2,3))
```

```python
# Experimented with __call__ function

class Test:
    def __call__(self):
        print("Object called like a function")

obj = Test()
obj()        
```

```python
# Class based decorator

class Decorator:
    def __init__(self,func):
        self.func = func

    def __call__(self):  
        print("Before function call")
        self.func() 
        print("After function call") 

@Decorator        
def hello():
    print("Hello World") 

def hello_2():
    print("Hello World Again!")


hello()

hello_2 = Decorator(hello_2)
hello_2()


```
